### Renaming the 'adr' column name to 'Average_daily_rate' for understanding the data well.

### Replacing null/missing values in columns(agent, children,company,country)

```python
hotel_data['children'].fillna(hotel_data.children.mean(), inplace = True) 
hotel_data['children'] = hotel_data['children'].astype(int)    
hotel_data['company'].fillna(0 , inplace = True)             
hotel_data['company'].astype(int)                            
hotel_data['agent'].fillna(0 , inplace = True)              
hotel_data['agent'].astype(int)                               
hotel_data['country'].fillna('Other',inplace = True)
```

### Removing duplicated rows

```python
hotel_data[hotel_data.duplicated()].shape
```

```python
hotel_data.drop_duplicates(inplace = True) 
```

### Now we can see we dont have any null values in the data .

```python
hotel_data.info()
```

```python
hotel_data.describe()
```

There are some rows with total number of adults, children or babies equal to zero. So we will remove such rows.

```python
hotel_data[hotel_data.adults + hotel_data.babies + hotel_data.children == 0].shape
```

```python
hotel_data.drop(hotel_data[hotel_data.adults + hotel_data.babies + hotel_data.children == 0].index, inplace = True)
```

```python
hotel_data
```

# Removing outliers and adding some columns

Adding a column

```python
hotel_data['kids'] = hotel_data.children + hotel_data.babies           
hotel_data['Full_stay'] = hotel_data.stays_in_weekend_nights + hotel_data.stays_in_week_nights
```

Removing an outlier from Average_daily_rate column

```python
plt.figure(figsize = (12,6))
sns.scatterplot(y = 'Average_daily_rate', x = 'Full_stay', data = hotel_data)
plt.show()
```

```python
hotel_data.drop(hotel_data[hotel_data['Average_daily_rate'] > 5000].index, inplace = True)
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(y = 'Average_daily_rate', x = 'Full_stay', data = hotel_data)
plt.show()
```

From the scatter plot we can see that as length of Full_stay increases the Average_daily_rate decreases. This means for longer stay, the better deal(discounts) for customer can be finalised.

Removing outliers from 'stays_in weekend_nights' and 'stays_in_week_nights'

```python
hotel_data.loc[hotel_data.stays_in_weekend_nights >= 5,'stays_in_weekend_nights']=  5 
hotel_data.loc[hotel_data.stays_in_week_nights >= 10,'stays_in_week_nights']= 10  
```

Adding new column

```python
hotel_data['Total_members'] = hotel_data.adults + hotel_data.children + hotel_data.babies
hotel_data.loc[hotel_data.Total_members > 20, 'Total_members'] = 20
```

```python
hotel_data.info()
```

Creating subset dataframes for 'City' and 'Resort' hotel data

```python
City_data =pd.DataFrame(hotel_data[hotel_data['hotel'] =='City Hotel'])
Resort_data =pd.DataFrame(hotel_data[hotel_data['hotel'] =='Resort Hotel'])
```

Firstly finding the correlation between the numerical data. Also, adding Full_stay and Total_members columns. So, adults, children, babies, stays_in_weekend_nights, stays_in_week_nights columns could be added.

```python
Corr_data = hotel_data[['lead_time','previous_cancellations','previous_bookings_not_canceled','booking_changes',
                        'days_in_waiting_list','Average_daily_rate','required_car_parking_spaces',
                        'total_of_special_requests','Full_stay','Total_members']]
```

```python
corr_mat = Corr_data.corr()
f, ax = plt.subplots(figsize=(12, 7))
sns.heatmap(corr_mat,vmax=0.8,annot = True,fmt='.2f');
```

1) Full_stay length and lead_time have slight correlation. This means that for longer hotel stays people generally plan little before the actual arrival date.

2) Average_daily_rate is slightly correlated with Total_members, which makes sense as more number of people means more revenue, therefore more Average_daily_rate.

```python
resort_guest = Resort_data['arrival_date_month'].value_counts().reset_index() 
resort_guest.columns=['month','no of guests']
resort_guest

city_guest = City_data['arrival_date_month'].value_counts().reset_index()
city_guest.columns=['month','no of guests']
city_guest

final_guest=resort_guest.merge(city_guest, on = 'month')                             
final_guest.columns=['month','no of guests in resort','no of guest in city hotel']
final_guest
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
final_guest['month'] = pd.Categorical(final_guest['month'], categories=months, ordered=True)


plt.figure(figsize =(15,8))
sns.lineplot(data=final_guest, x='month', y='no of guests in resort', marker ='^')
sns.lineplot(data=final_guest, x='month', y='no of guest in city hotel',marker ='^')
plt.legend(['Resort','City Hotel'])
plt.ylabel('Number of guest')
```

As seen above, the busiest months for both the hotels are May, June ,July, August.

# Waiting time

```python
grouped_by_hotel = hotel_data.groupby('hotel') #creating a DF which store groupby 'Hotel' rows
# waiting_df which stores mean of days in waiting list column
waiting_data = pd.DataFrame(grouped_by_hotel['days_in_waiting_list'].agg(np.mean).reset_index().rename(columns = {'days_in_waiting_list':'avg_waiting_period'}))
plt.figure(figsize = (8,8))
sns.barplot(x = waiting_data['hotel'], y = waiting_data['avg_waiting_period'] )
plt.show()
```

City hotel has significantly longer waiting time, hence City Hotel is much busier than Resort Hotel.

# Average daily rate & price

```python
plt.figure(figsize=(12,5))
sns.lineplot(data = hotel_data, x = 'arrival_date_month', y = 'Average_daily_rate', hue = 'hotel',sort =True, marker ='^')
plt.title('Average daily rate month wise',fontweight ='bold',size =20)
```

For resort hotels, the average daily rate is more expensive during august, july and september.

For city hotels, the average daily rate is more expensive during august, july, june and may.

```python
plt.figure(figsize=(12,5))

hotel_data['Average_Daily_rate_pp'] = hotel_data['Average_daily_rate'] / (hotel_data['adults'] + hotel_data['children']) 
Actual_guests = hotel_data.loc[hotel_data["is_canceled"] == 0]
Actual_guests['price'] = Actual_guests['Average_daily_rate'] * (Actual_guests['Full_stay'])
sns.lineplot(data = Actual_guests, x = 'arrival_date_month', y = 'price', hue = 'hotel', marker ='^')
plt.title('Average daily price per person',fontweight ='bold', size =20)
plt.show()
```

Prices of resort hotel are much higher during May,June,July,August,September

Prices of city hotel do not fluctuate that much.
